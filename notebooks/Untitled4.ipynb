```python
def label_encoding(df):
        internal_keys = df['internalStatus'].unique().tolist()
        internal_dict = {}
        for i, k in zip(range(0, 15), internal_keys):
            internal_dict[k] = i
        df['internalStatus_m'] = df['internalStatus'].map(internal_dict)
        df = df.drop(columns=['internalStatus'])
        return df
```
